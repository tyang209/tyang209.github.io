```python
import glob
import json
import re
import pandas as pd
import sqlalchemy
```

```python
json_files = glob.glob('/Users/tyang/Projects/maude_night/*/*.json')
output = []
for info in json_files:
    data = json.load(open(info))
    writer='no writer'
    m = re.search(r'(?:writ[t]en|writer:)(?: by| by:)? ([A-Za-z \.]*)(?=\n|\r|\m|,)',data['description'],flags=re.I)
    if m:
        writer = m.group(1)
    else:
        m2 = re.search(r'^(?:by|by:)? ([A-Za-z \.]*)(?=\n|\r|\m|,)', data['description'],flags=re.I)
        if m2:
            writer = m2.group(1)
    output.append((info.split('/')[5],data['fulltitle'],writer,data['webpage_url'],data['upload_date']))
        
df = pd.DataFrame(output)
df.columns = ['team','sketch_title','writer','link','upload_date']
df.to_csv('maude.csv',encoding='utf-8',index=False)
```

```python
def remove_writer(df):
    df['writer_clean'] = df['writer'].apply(lambda x: re.sub('(.*writ?ten|by?:\s|writer)','',x,flags=re.I))
    df['writer_clean'] = df['writer_clean'].apply(lambda x: re.sub('.*by[\s:]','',x,flags=re.I))
    df['writer_clean'] = df['writer_clean'].apply(lambda x: re.sub('pitche?d?.*','',x,flags=re.I).strip())
    return df
```
